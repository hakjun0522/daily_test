Q1_.0117 타이타닉 데이터셋(sns.load_dtaset('tatanic'))을 사용하여, 클래스(class)별 승객수를 나타내는 막대그래프를 생성하세요.
-각 막대위에 승객수 표

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 타이타닉 데이터셋 불러오기
titanic_data = sns.load_dataset('titanic')

# 클래스(class)별 승객 수 계산
class_counts = titanic_data['class'].value_counts()

# 막대 그래프 생성
plt.figure(figsize=(8, 6))
bar_plot = sns.barplot(x=class_counts.index, y=class_counts.values, palette='viridis')

# 각 막대위에 승객 수 표시
for p in bar_plot.patches:
    bar_plot.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                      textcoords='offset points')

# 그래프 제목과 축 레이블 추가
plt.title('Passenger Count by Class')
plt.xlabel('Class')
plt.ylabel('Passenger Count')

# 그래프 표시
plt.show()
```

Q2.0117 sns.load_dataset('flights')을 로드하고, 이를 피벗 테이블로 변홚아여 연도별 월간 승객수를 나타내세요.
이 데이터를 사용하여 히트맵을 생성하사ㅔ요. 생상맵은 자유롭게.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# flights 데이터셋 로드
flights_data = sns.load_dataset('flights')

# 피벗 테이블로 변환 (연도별 월간 승객수)
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers', aggfunc='sum')

# 히트맵 생성
plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(flights_pivot, cmap='coolwarm', annot=True, fmt='d', linewidths=.5)

# 그래프 제목과 축 레이블 추가
plt.title('Monthly Passenger Count by Year')
plt.xlabel('Year')
plt.ylabel('Month')

# 그래프 표시
plt.show()
```

Q3.0117 sns.load_dataset('penguins')을 사용하여 페어플룻(pairplot)을 생성하세요.종(species)에 따라 다른색상을 사용하세요.(hue사용)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# penguins 데이터셋 로드
penguins_data = sns.load_dataset('penguins')

# 페어플롯 생성 (종에 따라 다른 색상 사용)
pairplot = sns.pairplot(penguins_data, hue='species', palette='Set1')

# 그래프 제목 추가
plt.suptitle('Pairplot of Penguins by Species', y=1.02)

# 그래프 표시
plt.show()
```

Q4.0117. 타이타닉 데이터넷을 사용하여 객실등급(class)에 따른 탑승객의나이를 나타내는 박스 플롯과 바이올린 플롯을 같은 그래프
내에서 나란히 생성하세요 이를 통해 각 클래스 별 나이 분포의 차이를 비교 하세요.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 타이타닉 데이터셋 로드
titanic_data = sns.load_dataset('titanic')

# 그래프 크기 설정
plt.figure(figsize=(12, 6))

# 박스 플롯 생성
plt.subplot(1, 2, 1)
sns.boxplot(x='class', y='age', data=titanic_data, palette='Set2')
plt.title('Box Plot of Age by Class')

# 바이올린 플롯 생성
plt.subplot(1, 2, 2)
sns.violinplot(x='class', y='age', data=titanic_data, palette='Set2')
plt.title('Violin Plot of Age by Class')

# 그래프 표시
plt.show()
```

Q5.0117 펭귄 데이터셋을 사용하여, 각 종별로 섬(island)별 개체 수를 나타내는 카운트 플롯을 생성하세요.x축은 'island',hue는 'soecies'로 설정하세요.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 펭귄 데이터셋 로드
penguins_data = sns.load_dataset('penguins')

# 카운트 플롯 생성
sns.countplot(x='island', hue='species', data=penguins_data, palette='Set1')

# 그래프 제목 추가
plt.title('Count Plot of Penguins by Island and Species')

# 그래프 표시
plt.show()
```
