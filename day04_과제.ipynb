### ※ Q1_0116. Start 칼럼의 분포 및 이상치 여부를 탐색하는 Box Plot 및 이상치를 구하세요.
- 이상치를 구하는 일반적인 방법은 IQR(Interquartile Range)을 사용
- IQR은 데이터의 중간 50% 범위를 나타내며, 이상치는 일반적으로 IQR의 1.5배를 벗어나는 값을 가지는 데이터 포인트로 정의
- IQR은 데이터의 중간 50% 범위로, Q3(상위 75%)에서 Q1(하위 25%)을 뺀 값이며, IQR = Q3 - Q1로 계산된다.
- Upper Whisker(상단 이상치 경계)는 Q3 + (1.5 * IQR)로 계산되며, 이 경계를 넘어가는 값은 상단 이상치로 간주 됨.
- Lower Whisker(하단 이상치 경계)는 Q1 - (1.5 * IQR)로 계산되며, 이 경계를 넘어가는 값은 하단 이상치로 간주 됨.

```python
# Q1_답

Q1 = df['Start'].describe()['25%']
Q3 = df['Start'].describe()['75%']
iqr = Q3 - Q1
lower_whisker_check = Q1 - iqr * 1.5
upper_whisker_check = Q3 + iqr * 1.5
abv = df[(df['Start'] < lower_whisker_check) | (df['Start'] > upper_whisker_check)]
abv
```

### ※ Q3_0116. A,B 그룹의 Box Plot을 그리고 이상치를 구하시오.





```python
import seaborn as sns
import matplotlib.pyplot as plt

# Box Plot 그리기
sns.boxplot(x='species', y='length', data=fish_multi)
plt.show()

# 이상치 계산 및 출력
for species in fish_multi['species'].unique():    #unique 특정열의 모든 고유 값을 반환
    group = fish_multi[fish_multi['species'] == species]
    Q1 = group['length'].quantile(0.25)
    Q3 = group['length'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = group[(group['length'] < lower_bound) | (group['length'] > upper_bound)]
    print(f"이상치 for species {species}:\n{outliers}\n")
```

그려보세요.(예:선 그래프,막대그래프,산점도,히스토그)

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 생성


# 서브플롯 생성
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# 선 그래프
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
axes[0, 0].plot(x, y1, label='sin(x)', color='blue')
axes[0, 0].plot(x, y2, label='cos(x)', color='red')
axes[0, 0].set_title('sin,cos')


# 막대 그래프
categories = ['A', 'B', 'C']
values = [15, 30, 25]
axes[0, 1].bar(categories, values, color='green', alpha=0.8)
axes[0, 1].set_title('Bar Chart')

# 산점도
x_scatter = np.random.randn(30)
y_scatter = 3 * x_scatter + 2 + np.random.randn(30)
axes[1, 0].scatter(x_scatter, y_scatter, color='purple', marker='.')
axes[1, 0].set_title('Scatter Plot')

# 히스토그램
data_hist = np.random.randn(1000)
axes[1, 1].hist(data_hist, bins=50, color='orange', alpha=0.7)
axes[1, 1].set_title('Histogram')

plt.show()
```
